A graph is a mathematical structure used to represent relationships between pairs of elements. A graph consists of a set of nodes and a set of links connecting the nodes.

In data science, graphs can be used to represent and analyze relationships between different data elements. For example, graphs can be used to represent friendship relationships in a social network, collaboration relationships in a set of documents, or dependency relationships in a set of tasks. Graphs are a very useful tool in data science and can be used for many different purposes, including data visualization, pattern identification, and communication of results.

In this case the application we have sought is to show the relationships between users and the businesses to which they have written a review.

```python
import networkx as nx
import pandas as pd
from tqdm import tqdm
from IPython import display
```

## Data processing

```python
df_nighlife_PA_reviews = pd.read_csv('../data/processed/df_nightlife_PA_reviews.csv')
```

```python
# Create a list of the column names to include in the new DataFrame for the Network Model
columns_to_include_network = ['business_id', 'user_id', 'stars_x']
```

```python
# Use the df.loc[] method to select only the specified rows and columns
df_nightlife_network = df_nighlife_PA_reviews.loc[:, columns_to_include_network]
df_nightlife_network.to_csv('../data/processed/df_nightlife_network.csv', index=False)
```

```python
print('Data Shape:',df_nightlife_network.shape)
```

Given the immense amount of users and businesses that we have in our data, which involved the creation of more than 250,000 nodes and their corresponding edges, unfeasible for processing in the computer equipment we have, we have chosen to perform a massive criving and select only those most experienced users who have written more than 200 different reviews.

```python
# Group by user_id and count the number of reviews per user
reviews_per_user = df_nightlife_network.groupby('user_id').size()

# We filter the DataFrame to keep only those users who have made more than 200 review
df_network_200 = df_nightlife_network[df_nightlife_network['user_id'].isin(reviews_per_user[reviews_per_user > 200].index)]

print('Number of business:', df_network_200['business_id'].nunique())
print('Number of users:', df_network_200['user_id'].nunique())
print('Number of reviews:', len(df_network_200))
```

As a result of the criving we have obtained a total of 3 users who have made 788 reviews to 434 different businesses.

## Network design and creation

```python
# Create an empty graph
G_200 = nx.Graph()
for index, row in tqdm(df_network_200.iterrows()):
    # Add the nodes to the graph
    G_200.add_node(row['business_id'])
    G_200.add_node(row['user_id'])
    # Add the edge to the graph and assign the weight indicated in stars_x
    G_200.add_edge(row['business_id'], row['user_id'], weight=row['stars_x'])

G_nodes = G_200.number_of_nodes()
G_edges = G_200.number_of_edges()
print("Nodes = ", G_nodes, " Edges = ", G_edges)
```

The result of the network is 437 nodes corresponding to the 434 businesses and 3 users.

```python
# We save the network file
nx.write_gexf(G_200, "../data/processed/nightlife_network_200.gexf")
```

## Network analysis

```python
display.Image("../images/nightlife_network_200.png")
```

To visualize the graph, we have executed the gexf file with the Gephi program.

We have used the 'Force Atlas' distribution algorithm, parameterizing it to obtain the most appropriate visualization in order to explain the graph properly.

The larger nodes in red represent the 3 selected users, the nodes in blue represent the evaluated businesses.

The edges that go from the nodes of the users to the nodes of the businesses represent the score that has been introduced in the review, the color of this edge indicates the value of the score, being red a negative score 1 and gradually modifying its color up to dark green that indicates a very positive score 5.

We can see that there are several reddish edges that converge in the same business, this allows us to conclude that if several reviews from different users agree on a negative score, it is very likely that the experience in that business is not satisfactory. In the same way, it is possible to locate businesses in which very positive user reviews converge, leading to the conclusion that the experience in that business is positive.
